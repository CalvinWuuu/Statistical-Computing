首先我要先有隨機亂數，且隨機這邊首先是會有範圍限制，再來就是均勻分布！總不能常態分布吧？都分布在中間，那就是均勻分布！
我要在這 N 個範圍隨機撒下 1000 個點，還要亂數喔！

```python
# Generate N samples in X-Y plane from uniform distribution
import numpy as np
import matplotlib.pyplot as plt
N = 1000
x = np.random.uniform(0, 1, N)
y = np.random.uniform(0, 1, N)
plt.scatter(x, y)
plt.axis('equal')
plt.show()
```

```python
# Generate N samples in X-Y plane from uniform distribution
N = 10000
x = np.random.uniform(-1, 1, N)
y = np.random.uniform(-1, 1, N)

d = x**2 + y**2 # 10000 distances from origin
n = np.sum(d <= 1) # number of points within the unit circle 小於 1 給 1，大於 1 給 0，就看有沒有落在圓內
pi_estimate = 4 * n / N
print(f"Estimated value of π: {pi_estimate}")
```

我這個實驗想做 10000 次，然後我想畫個直方圖來看看！

```python
# Do the same experiment above for 100 times and plot a histogram of the results
num_experiments = 100
pi_estimates = []
N = 10000  # Number of samples in each experiment
for _ in range(num_experiments):
    x = np.random.uniform(-1, 1, N)
    y = np.random.uniform(-1, 1, N)
    d = x**2 + y**2
    n = np.sum(d <= 1)
    pi_estimate = 4 * n / N
    pi_estimates.append(pi_estimate)
plt.hist(pi_estimates, bins=30, edgecolor='black')
plt.axvline(np.pi, color='red', linestyle='dashed', linewidth=1)
plt.title('Histogram of π Estimates')
plt.xlabel('Estimated π Value')
plt.ylabel('Frequency')
plt.show()
```

老師做的：

```python
N = 100000
M = 10000 # number of monte carlo simulations
pi_estimate = np.zeros(M)

for i in range(M):
    x = np.random.uniform(-1, 1, N)
    y = np.random.uniform(-1, 1, N)
    d = x**2 + y**2
    n = np.sum(d <= 1)
    pi_estimate[i] = 4 * n / N

# draw histogram
plt.hist(pi_estimate, bins=30, edgecolor='black')
plt.axvline(np.pi, color='red', linestyle='dashed', linewidth=1)
plt.title('Histogram of π Estimates')
plt.xlabel('Estimated π Value')
plt.ylabel('Frequency')
plt.show()

# Mean estimate of pi
mean_pi = np.mean(pi_estimate)
print(f"Mean estimated value of π over {M} simulations: {mean_pi}")
```
